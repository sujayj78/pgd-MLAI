## Data Cleaning

The **weathersit** variable is an unordered categorical variable hence we create dummy variables for it be useful for model building

```python
df['weathersit'].value_counts()
```

```python
# Column names are set according to the first word definitions given in the data dictionary
weather_string = {1:'Clear',2:'Mist',3:'Light Snow',4:'Heavy Rain'}
df['weathersit'] = df['weathersit'].apply(lambda x: weather_string[x])
```

```python
weather = pd.get_dummies(df['weathersit'], drop_first = True)
weather
```

```python
df = pd.concat([df,weather], axis = 1)
df.drop(['weathersit'], axis = 1, inplace = True)
df.head(10)
```

The **season** variable is an unordered categorical variable hence we create dummy variables for it be useful for model building

```python
df['season'].value_counts()
```

```python
season_string = {1:'spring',2:'summer',3:'fall',4:'winter'}
df['season'] = df['season'].apply(lambda x: season_string[x])
```

```python
season = pd.get_dummies(df['season'], drop_first = True)
season
```

```python
df = pd.concat([df,season], axis = 1)
df.drop(['season'], axis = 1, inplace = True)
df.head(10)
```

Remove the **instant**, **dteday** variables as they are like indexes to the dataframe which won't be useful for analysis. 

Remove the **casual** and **registered** variables as they are similar to target variable which accounts for number of bikes rented

```python
df_clean = df.drop(['instant','dteday','casual','registered'],axis=1)
df_clean.head(10)
```

## Step 4: Splitting the Data into Training and Testing Sets


```python
np.random.seed(0)
df_train, df_test = train_test_split(df_clean, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
df_clean.columns
```

```python
df_clean.describe()
```

It can be noticed that the values for each variable have different ranges and scales, this can become inconvenient while building the model, hence we normalize all variables according to Min max scaling

```python
scaler = MinMaxScaler()
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables
num_vars = ['yr', 'mnth', 'holiday', 'weekday', 'workingday', 'temp', 'atemp',
       'hum', 'windspeed', 'cnt', 'Light Snow', 'Mist',
       'spring', 'summer', 'winter']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
df_train.head(10)
```

```python
df_train.describe()
```

Plot the heat map to show correlaetion of all variables to understand the relation between different variables

```python
plt.figure(figsize = (16, 10))
sns.heatmap(df_train.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

### Method for training and summarising multiple regression model using statsmodels

```python
def train_and_summarise(X_train,y_train):
    X_train_lm = sm.add_constant(X_train)
    lr = sm.OLS(y_train, X_train_lm).fit()
    print("params: ",lr.params)
    print(lr.summary())
    return lr
```

### Method for checking VIF


```python
def print_vif(X_train):
    # Create a dataframe that contains names of all the feature variables and their respective VIFs
    vif = pd.DataFrame()
    vif['Features'] = X_train.columns
    vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    print(vif)
    
```

### Adding all the variables to the model

```python
y_train = df_train.pop('cnt')
X_train = df_train
```

```python
# Fit the model for all variables and investigate the summary
lr_1 = train_and_summarise(X_train,y_train)
```

```python
# Print the VIF between all dependable variables
print_vif(X_train)
```

#### We see that atemp has very high p value and also VIF value, so we drop it

```python
X_train = X_train.drop('atemp', 1,)
```

```python
lr_2 = train_and_summarise(X_train,y_train)
```

```python
print_vif(X_train)
```

#### We see that mnth has very high p value and also 4rd highest VIF value, so we drop it


```python
X_train = X_train.drop('mnth', 1,)
```

```python
lr_3 = train_and_summarise(X_train, y_train)
```

```python
print_vif(X_train)
```

#### We see that hum has high VIF value, we drop it to see if performance improves


```python
X_train_tmp = X_train.drop('hum', 1,)
```

```python
lr_4 = train_and_summarise(X_train_tmp,y_train)
```

#### We can see that removing hum has dropped the adjusted r_2 value, hence we retain hum and try removing temp as it had next highest VIF value

```python
X_train_tmp = X_train.drop('temp', 1,)
```

```python
lr_5 = train_and_summarise(X_train_tmp,y_train)
```

#### We can see that removing temp has dropped the adjusted r_2 value, hence we retain temp

### Residual analysis

For the chosen model parameters, we plot the error between predicted train values and ground truth train values to see if a normal distribution is maintained

```python
X_train_lm = sm.add_constant(X_train)
y_train_pred = lr_3.predict(X_train_lm)
```

```python
fig = plt.figure()
sns.distplot((y_train - y_train_pred), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)        
```

###  Normal distribution is occuring, hence the residual analysis is successful and the model can be used on the test dataset

### Making predictions

```python
num_vars = ['yr', 'mnth', 'holiday', 'weekday', 'workingday', 'temp', 'atemp',
       'hum', 'windspeed', 'cnt', 'Light Snow', 'Mist',
       'spring', 'summer', 'winter']
df_test[num_vars] = scaler.transform(df_test[num_vars])
```

```python
# Seperate the dataset into X and y
y_test = df_test.pop('cnt')
X_test = df_test
```

```python
# Drop the methods which were rejected during model building
X_test = X_test.drop(['atemp','mnth'],axis=1)
```

```python
# Get predicted 'cnt' values
X_test_lm = sm.add_constant(X_test)
y_pred = lr_3.predict(X_test_lm)
```

```python
# The R squared value on test dataset is as follows
r2_score(y_test, y_pred)
```
